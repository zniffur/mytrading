```python
df3 = pd.read_csv('etf_info_vol20_clean_mach1.csv')
df3['endDate'] =  pd.to_datetime(df3['endDate'])
df3['startDate'] =  pd.to_datetime(df3['startDate'])
df3.drop(df3.columns[0] ,axis=1, inplace=True)
```

```python
def get_acc_dist(row):
    isin = row['isin']
    main_url = "https://www.justetf.com/it/etf-profile.html?query="+isin+"&groupField=index&from=search&isin="+isin
    req = requests.get(main_url)
    soup = BeautifulSoup(req.text, "html.parser")
    acc = soup.find_all("td",text="Accumulazione")
    dist = soup.find_all("td",text="Distribuzione")   
    if acc != []:
        #print(isin, 'Accumulazione')
        print(isin, acc)
        #print(main_url)
        return 'ACC'
    elif dist != []:
        #print(isin, 'Distribuzione')
        print(isin, dist)
        #print(main_url)
        return 'DIST'
    else:
        return '-'

    
etf_info = df3.copy()

etf_info['ACCDIST'] = etf_info.apply(get_acc_dist, axis=1)

```
